```python
### Preparing the workbook
import numpy as np
import pandas as pd
import pyfolio as pf
import talib 

%matplotlib inline
```

```python
### Reading 中证500 ETF（510500）
data = pd.read_excel('../../data/500ETF_510500_30m_20151028.xlsx', index_col='date', parse_dates=True)
price_close = np.array(data['close'])
macd, macdsignal, macdhist = talib.MACD(np.array(data['close']))

data['ret'] = (data['close'] / data['close'].shift(1) - 1)
data['macd'] = macd
data['macdsignal'] = macdsignal
data['macdhist'] = macdhist
```

```python

data['ret_strategy'] = data['ret'] * (data['macdhist'].shift(1) > 0)
(data['ret_strategy'][4100::]).cumsum().plot()
```
